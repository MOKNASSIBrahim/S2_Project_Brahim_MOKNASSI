```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg
def u0(x):
    if x>=-1 and x<=1:
        return 5
    else:
        return 0
```

# Test 1

```python
N=200
h=20/(N-1)
c=1
dt=0.05
T=5
X=np.linspace(-10,10,N+2)
def uex(t,x):
    return u0(x-c*t)




d=4*np.ones(N-2)
d=np.append(d,2)
d=np.insert(d,0,2)
d2=np.ones(N-1)
A1=np.diag(d2,1)
A2=np.diag(d2,-1)
A3=np.diag(d)
M=h/6*(A1+A2+A3)
```

```python
d=np.zeros(N-2)
d=np.append(d,1)
d=np.insert(d,0,-1)

d2=np.ones(N-1)
A1=np.diag(-d2,1)
A2=np.diag(d2,-1)
A3=np.diag(d)

Nm=c/2*(A1+A2+A3)

C=M-dt*Nm
B=np.linalg.inv(C)

```

```python
u=np.zeros(N+2)
for i in range(N+2):
    u[i]=u0(X[i])
# Tracé de la condition initiale
plt.plot(X,u,'-r',label='Exact Sol (t = 0)')
plt.plot(X,u,'-b',label='Approched Sol (t = 0)')
plt.legend()
plt.pause(0.1)



temps=0
unew=np.zeros(N)





TT=np.linspace(0,T,101)
E=np.zeros(101)


u=np.zeros(N)
for i in range(N):
    u[i]=u0(X[i])
    
j=0
while temps<=T:
    
    unew= B @ M @ u
    v = unew.copy()
    v=np.append(v,v[N-1])
    v=np.insert(v,0,v[0])

    
    
    u = unew.copy()
    
    # Courbes de u au cours du temps   
    if  j % 20 == 0 and j!=0:
        plt.plot(X, Uexact, '-r', label='Sol exacte (t = {:.2f})'.format(temps))
        plt.plot(X, v, '-b', label='Sol approchee (t = {:.2f})'.format(temps))
        plt.legend()
        plt.pause(0.1)
    
    
    
    temps+=dt
    
    
    Uexact=np.zeros(N+2)
    for i in range(N+2):
        Uexact[i]=uex(temps,X[i])
        
    e=np.sqrt(h)*linalg.norm((Uexact-v))
    E[j]=e
    j+=1
```

```python
plt.plot(TT,E,'-b',label="L'erreur en fonction du temps")
plt.legend()
plt.xlabel('Time')
plt.ylabel('Error')
plt.title('The error of the first test')
```
